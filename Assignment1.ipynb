**2. Variables of different data types**

```python
# Creating one variable of each data type and printing their type
a_str = "Hello World!!"
a_int = 102
a_float = 12.14
a_bool = True

print(type(a_str))
print(type(a_int))
print(type(a_float))
print(type(a_bool))
```

**3. Area & Perimeter of Rectangle**

```python
# Calculating area and perimeter of a rectangle with length 7 and width 5
l = 17
w = 5
a = l * w
p = 2 * (l + w)
print("Area:", a)
print("Perimeter:", p)
```

**5. List operations**

```python
# Creating a list of numbers, adding one, removing one, and sorting the list
numbers = [2,10, 3, 25, 7, 1]

numbers.append(15)  # Adding new number
numbers.remove(3)   # Removing a number
numbers.sort()      # Sorting list

print(numbers)
```

**8. Sets: union & intersection**

```python
# Creating two sets of colors and finding union and intersection
s1 = {"red", "blue", "green"}
s2 = {"yellow", "blue", "black"}

print("Union:", s1.union(s2))
print("Intersection:", s1.intersection(s2))
```

***9. Function: print first & last character***

```python
# Function that prints the first and last character of a string
def show_first_last(text):
    print("First character:", text[0])
    print("Last character:", text[-1])

show_first_last("lovely")
```
